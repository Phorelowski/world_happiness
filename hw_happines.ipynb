```python
df_gdp = df9.sort_values('GDP per capita', ascending=False)
df_sup = df9.sort_values('Social support', ascending=False)
df_hp = df9.sort_values('Healthy life expectancy', ascending=False)
df_freedom = df9.sort_values('Freedom to make life choices', ascending=False)
df_geno = df9.sort_values('Generosity', ascending=False)
df_cor = df9.sort_values('Perceptions of corruption', ascending=False)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(2, 3, 1)
sns.barplot(y= df_gdp['Country or region'][:10],x= df_gdp['GDP per capita'][:10], color='green')
plt.xlabel('GDP per capita')
plt.ylabel('Country')

plt.subplot(2, 3, 2)
sns.barplot(y= df_sup['Country or region'][:10],x= df_sup['Social support'][:10], color='green')
plt.xlabel('Social support')
plt.ylabel('Country')

plt.subplot(2, 3, 3)
sns.barplot(y= df_hp['Country or region'][:10],x= df_hp['Healthy life expectancy'][:10], color='green')
plt.xlabel('Healthy life expectancy')
plt.ylabel('Country')

plt.subplot(2, 3, 4)
sns.barplot(y= df_freedom['Country or region'][:10],x= df_freedom['Freedom to make life choices'][:10], color='green')
plt.xlabel('Freedom')
plt.ylabel('Country')

plt.subplot(2, 3, 5)
sns.barplot(y= df_geno['Country or region'][:10],x= df_geno['Generosity'][:10], color='green')
plt.xlabel('Generosity')
plt.ylabel('Country')

plt.subplot(2, 3, 6)
sns.barplot(y= df_cor['Country or region'][:10],x= df_cor['Perceptions of corruption'][:10], color='green')
plt.xlabel('Perceptions of corruption')
plt.ylabel('Country')
plt.subplots_adjust(wspace=0.5, hspace=0.5)
```
